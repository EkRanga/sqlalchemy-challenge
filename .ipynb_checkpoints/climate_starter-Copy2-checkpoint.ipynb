# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from datetime import datetime, timedelta
from sqlalchemy import func, and_
from sqlalchemy import inspect
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
Base.prepare(engine, reflect=True)

# reflect the tables

inspector = inspect(engine)
inspector.get_table_names()
```

```python
# View all of the classes that automap found
def getMetaData(TableName):
    print(TableName)
    print("--------------") 
    columns = inspector.get_columns(TableName)
    for c in columns:
        print(c['name'], c["type"])
    print("--------------")    

table_names = Base.classes.keys()
for table_name in table_names:
    getMetaData(table_name)
    
```

```python
# Save references to each table
session = Session(engine)
Measurement = Base.classes.measurement
Station = Base.classes.station
# Example query: select all rows from the 'measurement' table
measurement_results = session.query(Measurement).all()

# Example query: select all rows from the 'station' table
station_results = session.query(Station).all()
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
latest_year_query = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
latest_year_query
```

```python
latest_date_query = session.query(func.max(Measurement.date)).scalar()
latest_date = datetime.strptime(latest_date_query, '%Y-%m-%d')  # Convert to datetime object
# latest_date
# #Step 2: Calculate the date that is exactly 12 months before the most recent date
one_year_ago = latest_date - timedelta(days=365)

# # Step 3: Query the database for precipitation data within the date range
last_12_months_prcp = session.query(Measurement.date, Measurement.prcp).\
    filter(and_(Measurement.date >= one_year_ago.strftime('%Y-%m-%d'),
                Measurement.date <= latest_date.strftime('%Y-%m-%d')))

df=pd.DataFrame(last_12_months_prcp.all())
df["date"] = pd.to_datetime(df["date"])


# Create figure and plot space
fig, ax = plt.subplots(figsize=(6, 4))

# Add x-axis and y-axis
ax.bar(df["date"],
       df['prcp'],
       width=6.0)

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Precipitation (inches)",
       title="Daily Precipitation")

ax.set_xticklabels(ax.get_xticklabels(), rotation="vertical", fontsize=10)

# Define the date format
date_form = DateFormatter("%Y-%m-%d")
ax.xaxis.set_major_formatter(date_form)

# Ensure a major tick for each week using (interval=1) 
# ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))

plt.show()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number of stations in the dataset
query = "SELECT COUNT(*) AS total_stations FROM station"

# Execute the query and fetch the result into a Pandas DataFrame
result = engine.execute(query)

# Fetch the result
total_stations = result.scalar()

# Display the result
print(total_stations)
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
# List the stations and their counts in descending order.

station_df=pd.read_csv("Resources/hawaii_measurements.csv")


# Group by station and count the rows
result = station_df.groupby('station').size().reset_index(name='row_count')

# Sort the result in descending order based on row count
result = result.sort_values(by='row_count', ascending=False)
result_tuples = list(result.itertuples(index=False, name=None))
# Display the results
# Display each tuple on a new line
for station, row_count in result_tuples:
    print(f"('{station}', {row_count})")
```

```python
file_path = 'hawaii_measurements.csv'  # Update this with the correct path to your file
df = pd.read_csv("Resources/hawaii_measurements.csv")

# Filter the DataFrame for the most active station (USC00519281)
most_active_station_data = df[df['station'] == 'USC00519281']

# Extract the temperatures from the 'tobs' column
temperatures = most_active_station_data['tobs']

# Calculate the lowest, highest, and average temperature
min_temperature = temperatures.min()
max_temperature = temperatures.max()
avg_temperature = temperatures.mean()

print(f"Most Active Station: USC00519281")
print(f"Lowest Temperature: {min_temperature}°C")
print(f"Highest Temperature: {max_temperature}°C")
print(f"Average Temperature: {avg_temperature}°C")
```

```python
# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Calculate the date 12 months ago from the latest date in the dataset
latest_date = df['date'].max()
one_year_ago = latest_date - timedelta(days=365)

# Filter the data for the most active station and the last 12 months
filtered_data = df[(df['station'] == 'USC00519281') & (df['date'] >= one_year_ago)]
filtered_data
# Plot the results as a histogram
plt.figure(figsize=(6, 5))
plt.hist(filtered_data['tobs'], bins=12)

plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```
